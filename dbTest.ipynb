```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from ipywidgets import Layout, interact
from IPython.display import display

dat = pd.read_csv('Spectro_14NH3.csv',header=2)
dat = dat.replace('-',np.nan)
dat.iloc[:,0] = dat.iloc[:,0].astype(float)
dat.iloc[:,2] = dat.iloc[:,2].astype(float)
dat.iloc[:,5] = dat.iloc[:,5].astype(float)
dat.iloc[:,6] = dat.iloc[:,6].astype(float)
dat.iloc[:,9] = dat.iloc[:,9].astype(float)
dat.iloc[:,10] = dat.iloc[:,10].astype(float)
dat.iloc[:,13] = dat.iloc[:,13].astype(float)
dat.head()
```

```python
def FindPumpFreq(FreqR, db):
    f = db[(((db['P Freq (wn)'] >= FreqR[0]) & (db['P Freq (wn)'] <= FreqR[1])) |
            ((db['Q Freq (wn)'] >= FreqR[0]) & (db['Q Freq (wn)'] <= FreqR[1])) | 
             ((db['R Freq (wn)'] >= FreqR[0]) & (db['R Freq (wn)'] <= FreqR[1])))]
    
    return f

test = FindPumpFreq([970,980],dat)
test.head()
    
```

```python
def FindTHzFreq(FreqR, db):
    f = db[((db['Freq (GHz)'] >= FreqR[0]) & (db['Freq (GHz)'] <= FreqR[1]))]
    return f

test = FindTHzFreq([5000,7000],dat)
test2 = FindPumpFreq([600,650],test)
```

```python
out1 = widgets.Output()
out1
```

```python
minTHz = dat['Freq (GHz)'].min()
maxTHz = dat['Freq (GHz)'].max()
minPump = 1e7
maxPump = 0
for i in range(3):
    tmpMin = dat.iloc[:,2+4*i].min()
    tmpMax = dat.iloc[:,2+4*i].max()
    if tmpMin < minPump:
        minPump = tmpMin
    if tmpMax > maxPump:
        maxPump = tmpMax

def on_button_clicked(b):
    tmp = FindPumpFreq(PumpRange.value,dat)
    tmp2 = FindTHzFreq(THzRange.value,tmp)
    out1.clear_output()
    #print(tmp2)
    with out1:
        display(tmp2)
    return(tmp2)

THzRange = widgets.FloatRangeSlider(
    value=[minTHz, maxTHz],
    min=minTHz,
    max=maxTHz,
    step=0.1,
    description='Laser (GHz):',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    layout=Layout(width='500px'),
)

PumpRange = widgets.FloatRangeSlider(
    value=[minPump, maxPump],
    min=minPump,
    max=maxPump,
    step=0.1,
    description='Pump (wn):',
    disabled=False,
    continuous_update=False,
    orientation='horizontal',
    readout=True,
    readout_format='.1f',
    layout=Layout(width='500px'),
)

SearchButton = widgets.Button(
    description='Find Frequencies',
    disabled=False,
    button_style='success', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Find Frequencies',
    icon=''
)

SearchButton.on_click(on_button_clicked)

display(THzRange,PumpRange, SearchButton)

out = widgets.Output()
```
